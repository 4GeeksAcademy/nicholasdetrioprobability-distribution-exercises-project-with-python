## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Exercise 1

Write the above functions but for both normal and uniform distributions.

```python
#NORMAL DISTRIBUTION

from scipy.stats import norm
from scipy.stats import binom

import matplotlib.pyplot as plt
import numpy as np

#DENSITY "I'M YOUR DENSITY, I MEAN DESTINY" - BTTF
def pdf_norm(x, mean = 0, standard_deviation = 1):

    result = norm.pdf(x, loc = mean, scale = standard_deviation)
        #for x the function will return the probability density at that specific point.

    return result

#CUMULATIVE
def cdf_norm(q,mean=0, standard_deviation=1):

    result = norm.cdf(q,loc=mean, scale=standard_deviation) 
        # The k parameter represents the quantile(s) at which you want to evaluate the cumulative distribution function (CDF).
    
    return result

#QUANTILE 
def ppf_norm(p,mean=0, standard_deviation=1):

    result = norm.ppf(q=p, loc=mean, scale=standard_deviation)
    # The q parameter represents the cumulative probability up to which you want to find the corresponding point(s) in the inverse CDF.
    
    return result


```

```python
# OUTPUT GRAPHS NORMAL DISTRIBUTION

# np.random.seed(42)

mean=0
standard_deviation=1
data = np.arange(-10, 10, 0.01)

pdf = pdf_norm(data, mean=mean, standard_deviation = standard_deviation)
cdf = cdf_norm(data, mean=mean, standard_deviation = standard_deviation)
ppf = ppf_norm(data, mean=mean, standard_deviation = standard_deviation)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pdf, "r-")
axis[0].set_title('PDF')

axis[1].plot(data, cdf, "y-")
axis[1].set_title('CDF')

axis[2].plot(data, ppf, color="purple")
axis[2].set_title('PPF')

plt.tight_layout()
plt.show()
```

```python
#UNIFORM DISTRIBUTION

from scipy.stats import norm
from scipy.stats import binom
from scipy.stats import uniform

import matplotlib.pyplot as plt
import numpy as np

#DENSITY "I'M YOUR DENSITY, I MEAN DESTINY" - BTTF
def pdf_uniform(x, lowest = 0, highest = 1):
    result = uniform.pdf(x, loc = lowest, scale = (highest-lowest))
        #for x the function will return the probability density at that specific point.
    return result

#CUMULATIVE
def cdf_uniform(q,lowest = 0, highest = 1):
    result = uniform.cdf(q,loc = lowest, scale = (highest-lowest)) 
        # The k parameter represents the quantile(s) at which you want to evaluate the cumulative distribution function (CDF).
    return result

#QUANTILE 
def ppf_uniform(p,lowest = 0, highest = 1):
    result = uniform.ppf(q=p, loc = lowest, scale = (highest-lowest))
    # The q parameter represents the cumulative probability up to which you want to find the corresponding point(s) in the inverse CDF.
    return result

```

```python
# OUTPUT GRAPHS UNIFORM DISTRIBUTION

# np.random.seed(42)

lowest = 0
highest = 1
data = np.arange(-10, 10, 0.01)

pdf = pdf_uniform(data, lowest = lowest, highest = highest)
cdf = cdf_uniform(data, lowest = lowest, highest = highest)
ppf = ppf_uniform(data, lowest = lowest, highest = highest)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pdf, "r-")
axis[0].set_title('PDF')

axis[1].plot(data, cdf, "y-")
axis[1].set_title('CDF')

axis[2].plot(data, ppf, color="purple")
axis[2].set_title('PPF')

plt.tight_layout()
plt.show()
```
